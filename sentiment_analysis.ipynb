```python
import pandas as pd
from openai import OpenAI
import os
```

```python
# Set your OpenAI API key
OpenAI.api_key = os.environ["OPENAI_API_KEY"]
```

```python
# Function to get sentiment analysis from OpenAI
def get_sentiment(text):
    client = OpenAI()

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "system",
                "content": [
                    {
                        "type": "text",
                        "text": "You will be provided with a text from a research journal discussing chatgpt in nursing, and your task is to classify its sentiment as positive, neutral, or negative."
                    }
                ]
            },
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": text
                    }
                ]
            }
        ],
        temperature=0,
        max_tokens=256,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0
    )

    # Extract the sentiment from the response
    return response.choices[0].message.content
```

```python
# Load the text to analyze
data = pd.read_excel('scraped_text_v1.xlsx')
data.head()
```

```python
# Apply the sentiment analysis function to the text column
data['Sentiment'] = data['Text'].apply(get_sentiment)

# Display the results
print(data[['Text', 'Sentiment']])
```
