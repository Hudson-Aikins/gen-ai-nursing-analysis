```python
# Import Packages
import fitz  # PyMuPDF
import pandas as pd
import os
```

```python
# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text
```

```python
# Read the Excel file
excel_path = "/Users/hudsonaikins/Desktop/Research-Analysis/Ultimate_Comprehensive_Research_Articles.xlsx"
df = pd.read_excel(excel_path)
df
```

```python
# Specify the directory containing the PDFs
pdf_directory = "/Users/hudsonaikins/Desktop/Research-Analysis/Research-Citations-2"  # Update this path with the correct directory

# List all PDF files in the directory
pdf_files = [f for f in os.listdir(pdf_directory) if f.endswith('.pdf')]

# Create a list to store the text extracted from each PDF
texts = []

# Iterate through all PDF files and extract text
for pdf_filename in pdf_files:
    pdf_path = os.path.join(pdf_directory, pdf_filename)
    text = extract_text_from_pdf(pdf_path)
    texts.append({'Filename': pdf_filename, 'Text': text})

# Convert the list to a DataFrame
df_texts = pd.DataFrame(texts)
```

```python
df_texts
```

```python
# Add the extracted text to the original DataFrame
df_combined = pd.merge(df, df_texts, on="Filename", how="inner")
```
